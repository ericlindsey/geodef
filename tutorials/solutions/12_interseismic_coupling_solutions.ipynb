## 2–3. Down-dip recovery and plate-rate bias

Deep coupling is spatially smeared. A 10% high plate rate makes a fixed
backslip estimate appear about 9% less coupled, while moment-deficit rate from
backslip itself is unchanged.

## 4. Raw versus plate coordinates

For a backslip rake of -90°, positive plate-parallel rate maps to negative raw
dip slip. Both representations must predict identical velocity.


```python
x = np.linspace(-80.0, 80.0, 321)
V = 40.0
profiles = {D: V / np.pi * np.arctan(x / D) for D in (5.0, 15.0, 30.0)}
plate_rate = 0.05
backslip = np.array([0.01, 0.03, 0.05])
coupling = backslip / plate_rate
biased_coupling = backslip / (1.1 * plate_rate)
ss, ds = geodef.slip.from_rake(backslip, rake_degrees=-90.0)
assert np.allclose(ss, 0.0) and np.all(ds <= 0.0)
print("coupling:", coupling)
print("with plate rate 10% high:", biased_coupling)
print("profile center gradients:", {D: np.gradient(v, x)[x.size//2] for D, v in profiles.items()})
```

## Interpretation and alternatives

Coupling inherits uncertainty in both numerator and denominator; reporting only inversion covariance is incomplete.
