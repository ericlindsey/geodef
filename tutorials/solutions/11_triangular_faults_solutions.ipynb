## 4. A bent join

Whether a Laplacian connects across a segment boundary is a prior decision. A
physical tear may require removing that adjacency; a smooth bend may not.


```python
frame = geodef.LocalFrame(34.0, -118.0)
nodes = np.array([
    [-10_000.0, -5_000.0, -3_000.0], [0.0, -5_000.0, -4_000.0], [10_000.0, -5_000.0, -5_000.0],
    [-10_000.0, 0.0, -6_000.0], [0.0, 0.0, -7_000.0], [10_000.0, 0.0, -8_000.0],
    [-10_000.0, 5_000.0, -9_000.0], [0.0, 5_000.0, -10_000.0], [10_000.0, 5_000.0, -11_000.0],
])
triangles = np.array([[0,1,3],[1,4,3],[1,2,4],[2,5,4],[3,4,6],[4,7,6],[4,5,7],[5,8,7]])
fault = geodef.Fault.from_triangles(nodes, frame=frame, triangles=triangles)
amplitude = np.ones(fault.n_patches)
ss_rake, ds_rake = geodef.slip.from_rake(amplitude, rake_degrees=45.0)
ss_az, ds_az = geodef.slip.from_azimuth(amplitude, azimuth_degrees=90.0, fault_strike_degrees=fault.strike)
print("component difference norm:", np.linalg.norm(ss_rake-ss_az) + np.linalg.norm(ds_rake-ds_az))
print("Laplacian shape:", fault.laplacian.shape)
```

## Interpretation and alternatives

The nonzero direction difference is expected: one local rake is not one geographic azimuth on variable-strike triangles.
