```python
import matplotlib.pyplot as plt
import numpy as np

import geodef

rng = np.random.default_rng(0)
```

## 1–2. Correlation length and wrong covariance

Effective independent observations decrease monotonically as the correlation
length grows. Assuming too little correlation overstates information; assuming
too much discards useful short-scale contrast.

## 3. Exponential versus Gaussian

The Gaussian model is smoother near zero lag and decays faster in its tail at
the same nominal length.

## 4. Two scales

Covariances add. The long scale controls broad modes; the short scale controls
local differences, with influence set by each sill.


```python
lon, lat = np.meshgrid(np.linspace(-118.2, -117.8, 12), np.linspace(33.9, 34.1, 5))
lon, lat = lon.ravel(), lat.ravel()
effective = []
for length in (100.0, 2_000.0, 10_000.0, 50_000.0):
    C = geodef.data.spatial_covariance(
        lon, lat, sill=1e-4, correlation_length=length, nugget=1e-5
    )
    effective.append((np.trace(C) ** 2) / np.trace(C @ C))
C_short = geodef.data.spatial_covariance(lon, lat, sill=5e-5, correlation_length=2_000.0, nugget=1e-6)
C_long = geodef.data.spatial_covariance(lon, lat, sill=5e-5, correlation_length=20_000.0, nugget=0.0)
np.linalg.cholesky(C_short + C_long)
print("effective N:", effective)
```

## Interpretation and alternatives

Positive Cholesky factorization verifies numerical validity, not scientific adequacy of the covariance model.
