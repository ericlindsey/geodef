```python
import matplotlib.pyplot as plt
import numpy as np

import geodef

rng = np.random.default_rng(0)
```

## 1. Pack and unpack a fake blocked vector

Packing concatenates whole components; it does not interleave them. Unpacking
uses the known midpoint and should reproduce both inputs exactly.

## 2. Vectorize several slopes

A `(4, 1)` slope column broadcasts against a `(20,)` x row to produce four
curves. No Python loop is needed.

## 3. Convert a geographic table

Column selection changes latitude/longitude order; only depth is multiplied by
1,000. The operation should preserve row identity.

## 4–5. Repair invalid calls and labels

Validation messages name negative depth, dip outside `[0, 90]`, and unmatched
array lengths. Fix the source field rather than suppressing the exception. A
defensible comparison uses shared limits and labels every quantity and unit.


```python
strike = np.array([1.0, 2.0, 3.0])
dip = np.array([4.0, 5.0, 6.0])
blocked = geodef.slip.pack(strike, dip)
strike_check, dip_check = geodef.slip.unpack(blocked)
assert np.array_equal(strike, strike_check) and np.array_equal(dip, dip_check)

x = np.linspace(0.0, 1.0, 20)
slopes = np.array([0.5, 1.0, 2.0, 4.0])[:, np.newaxis]
curves = slopes * x + 1.0
assert curves.shape == (4, 20)

lat_lon_depth_km = np.array([[34.0, -118.0, 5.0], [34.1, -117.9, 7.0]])
lon_lat_depth_m = lat_lon_depth_km[:, [1, 0, 2]].copy()
lon_lat_depth_m[:, 2] *= 1_000.0
print(blocked, curves.shape, lon_lat_depth_m)
```

## Interpretation and alternatives

The key habit is to verify scientific meaning with shape and units, not merely that an operation ran.
