```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Variables and units

Store each fact separately so it can be reused in later calculations. An
f-string combines the values into readable output while keeping the units
visible.

```python
sample_name = "granite"
depth_km = 6.4
density_kg_m3 = 2710
message = f"The {sample_name} sample came from {depth_km} km depth. "
message += f"Its density is {density_kg_m3} kg/m³."
print(message)
```

## 2. Lists and slices

Python counts from zero. The stop in a slice is excluded, so `[:2]` selects
positions 0 and 1. A negative start counts from the end.

```python
temperatures_c = [12.5, 14.0, 17.2, 15.8, 13.1]
print(temperatures_c[:2])
print(temperatures_c[-2:])
print(temperatures_c[2])
```

## 3. A condition

Check the highest threshold first. If `magnitude >= 6` is false, the next branch
checks whether the value is at least 4. Anything remaining must be below 4.

```python
magnitude = 5.3
if magnitude >= 6:
    category = "large"
elif magnitude >= 4:
    category = "moderate"
else:
    category = "small"
print(category)
```

## 4. A loop and a function

One millimeter is one thousandth of a meter. The function expresses that rule
once, and the loop supplies each measurement in turn.

```python
def millimeters_to_meters(distance_mm):
    return distance_mm / 1000
```

```python
for distance_mm in [2, 5, 12]:
    print(millimeters_to_meters(distance_mm))
```

## 5. Array selection

NumPy calculates the mean over all five values. The comparison produces a
Boolean array, which is then used to retain only entries greater than 3.

```python
measurements = np.array([1.2, 4.8, 3.1, 7.5, 2.0])
print(np.mean(measurements))
print(measurements[measurements > 3])
```

## 6. Vectorization and broadcasting

Adding `np.newaxis` turns the three slopes into a `(3, 1)` column. The six x
positions then broadcast across each row, giving one row per slope.

```python
slopes = np.array([1.0, 2.0, 4.0])[:, np.newaxis]
x_values = np.arange(6.0)
lines = slopes * x_values + 0.5
print(lines, lines.shape)
```

## 7. Plotting

Row 1 is the second line and therefore has slope 2. Both axes identify a
quantity and a unit; the title identifies which line is shown.

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x_values, lines[1])
ax.set(xlabel="Distance (km)", ylabel="Value (m)",
       title="Line with slope 2 m/km");
```

## 8. Matrix multiplication

The matrix has three rows, so it produces three predictions. Its two columns
combine with the two model values. The output shape `(3,)` therefore means one
predicted number for each row.

```python
design = np.array([[1.0, 0.0], [1.0, 1.0], [1.0, 2.0]])
parameters = np.array([3.0, 0.4])
predictions = design @ parameters
print(predictions, predictions.shape)
```

## 9. A linear solve

`solve` finds the two unknown values. Multiplying the matrix by that solution
should reproduce `b`; `np.allclose` confirms equality within floating-point
roundoff.

```python
coefficients = np.array([[3.0, 1.0], [1.0, 2.0]])
outcomes = np.array([9.0, 8.0])
solution = np.linalg.solve(coefficients, outcomes)
print(solution, coefficients @ solution)
assert np.allclose(coefficients @ solution, outcomes)
```

## 10. Least squares

The first column of ones represents the line's intercept, and the second column
contains x. The four observations do not lie on one exact line, so the residuals
are small rather than all zero.

```python
x_values = np.array([0.0, 1.0, 2.0, 3.0])
data = np.array([1.0, 2.2, 2.9, 4.1])
design = np.column_stack([np.ones(x_values.size), x_values])
parameters, *_ = np.linalg.lstsq(design, data, rcond=None)
```

```python
predictions = design @ parameters
differences = data - predictions
print("intercept and slope:", parameters)
print("predictions:", predictions)
print("differences:", differences)
```
