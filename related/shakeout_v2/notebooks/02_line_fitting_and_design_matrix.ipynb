# 02: Line Fitting and the Design Matrix

## Learning objectives
- Build a design matrix for $y = mx + b$.
- Solve for slope/intercept with least squares.
- Interpret residuals and basic parameter uncertainty.

## Model form
For each observation $(x_i, y_i)$:

$$
y_i = mx_i + b + e_i
$$

In matrix form:

$$
\mathbf{y} = G\mathbf{m} + \mathbf{e}, \quad
G = \begin{bmatrix}
x_1 & 1 \\
x_2 & 1 \\
\vdots & \vdots \\
x_N & 1
\end{bmatrix}, \quad
\mathbf{m} = \begin{bmatrix} m \\ b \end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (7, 4)
plt.rcParams['axes.grid'] = True

rng = np.random.default_rng(7)
```

```python
# Synthetic data with known true model
m_true = 1.8
b_true = -0.7

x = np.linspace(0, 10, 30)
sigma = 0.8
noise = rng.normal(0.0, sigma, size=x.size)
y = m_true * x + b_true + noise

fig, ax = plt.subplots()
ax.scatter(x, y, s=20, label='Noisy data')
ax.plot(x, m_true * x + b_true, 'k--', label='True line')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Synthetic data for line fitting')
ax.legend()
plt.show()
```

```python
# Build design matrix
G = np.column_stack([x, np.ones_like(x)])

# Least-squares estimate
m_est, residual_sum, rank, singular_vals = np.linalg.lstsq(G, y, rcond=None)
slope_est, intercept_est = m_est

print('Estimated slope     :', round(float(slope_est), 4))
print('Estimated intercept :', round(float(intercept_est), 4))
print('Matrix rank         :', rank)
```

```python
y_pred = G @ m_est
residuals = y - y_pred

fig, ax = plt.subplots()
ax.scatter(x, y, s=20, label='Data')
ax.plot(x, y_pred, 'r', label='Least-squares fit')
ax.plot(x, m_true * x + b_true, 'k--', label='True line')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Line fit result')
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.scatter(x, residuals, s=20)
ax.axhline(0.0, color='k', linewidth=1)
ax.set_xlabel('x')
ax.set_ylabel('Residual y - y_pred')
ax.set_title('Residuals')
plt.show()
```

## Optional uncertainty estimate
If data noise is i.i.d. with variance $\sigma^2$, an estimate for parameter covariance is:

$$
C_m \approx s^2 (G^TG)^{-1}, \quad
s^2 = \frac{\sum r_i^2}{N-p}
$$

where $p=2$ parameters for line fitting.

```python
N = len(y)
p = 2
s2 = np.sum(residuals**2) / (N - p)
Cm = s2 * np.linalg.inv(G.T @ G)
std_m = np.sqrt(np.diag(Cm))

print('Std(slope)    ~', round(float(std_m[0]), 4))
print('Std(intercept)~', round(float(std_m[1]), 4))
```
